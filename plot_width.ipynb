```python
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load your CSV file
df = (pd.read_csv(r"C:\Users\Paolo\OneDrive\Desktop\Thesis\Mycos_old_data (1)\Mycos_old_data\2014\20140520\merged_widths_200_Ok.csv", header=None,skiprows=1).T).astype(float)
raw_labels = df.iloc[:, -1]
features = df.iloc[:, :-1]
```

```python
for i in range(features.shape[0]):
    plt.plot(features.columns, features.iloc[i], color='blue', alpha=0.3)


plt.title("Cucumber Width at Position 50")
plt.xlabel("Sample Index")
plt.ylabel("Width")
plt.show()
```

```python
# Separate features by label
infected = features[raw_labels == 1]
non_infected = features[raw_labels == 0]

# Plot
plt.figure(figsize=(12, 6))

# Non-infected cucumbers
for i in range(non_infected.shape[0]):
    plt.plot(non_infected.columns, non_infected.iloc[i], color='green', alpha=0.3, label='Non-infected' if i == 0 else "")

# Infected cucumbers
for i in range(infected.shape[0]):
    plt.plot(infected.columns, infected.iloc[i], color='red', alpha=0.3, label='Infected' if i == 0 else "")

plt.title("Cucumber Width Profiles by Infection Status dis*skeleton of SAM images 20140520 data")
plt.xlabel("distance from the tip")
plt.ylabel("Width")
plt.legend()
plt.show()
```
